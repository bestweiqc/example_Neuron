```python
from neuron import h
h.load_file('stdrun.hoc')
soma = h.Section(name='soma')
soma.insert('hh')
if w_Nat.value:
    soma.insert('na3')
    inf_nat = h.Vector()
    tau_nat = h.Vector()
    inf_nat.record(soma(0.5)._ref_minf_na3)
    tau_nat.record(soma(0.5)._ref_mtau_na3)
if w_Nap.value:
    soma.insert('nap')
    inf_nap = h.Vector()
    tau_nap = h.Vector()
    inf_nap.record(soma(0.5)._ref_minf_nap)
    tau_nap.record(soma(0.5)._ref_mtau_nap)
if w_H.value:
    soma.insert('hd')
    inf_h = h.Vector()
    tau_h = h.Vector()
    inf_h.record(soma(0.5)._ref_linf_hd)
    tau_h.record(soma(0.5)._ref_taul_hd)
#if w_Leak.value:
#    soma.insert('leak')
if w_Kdr.value:
    soma.insert('kdr')
    inf_kdr = h.Vector()
    tau_kdr = h.Vector()
    inf_kdr.record(soma(0.5)._ref_ninf_kdr)
    tau_kdr.record(soma(0.5)._ref_taun_kdr)
if w_Cadyn.value:
    soma.insert('cadyn')
    inf_cadyn = h.Vector()
    tau_cadyn = h.Vector()
    inf_cadyn.record(soma(0.5)._ref_uinf_cadyn)
    tau_cadyn.record(soma(0.5)._ref_utau_cadyn)
#if w_Sahp.value:
    #soma.insert('sAHP')
if w_Km.value:
    soma.insert('im')
    inf_km = h.Vector()
    tau_km = h.Vector()
    inf_km.record(soma(0.5)._ref_ninf_im)
    tau_km.record(soma(0.5)._ref_taun_im)
if w_Kap.value:
    soma.insert('kap')
if w_Capool.value:
    soma.insert('capool')

v0_vec = h.Vector()
t_vec = h.Vector()
v0_vec.record(soma(0.5)._ref_v)
t_vec.record(h._ref_t)


def activemodel(diam,cm,L,el,gl,gnap,KNap,VhalfNap,gh,KH,VhalfH,gnat,KNat,
                gkdr,VhalfKdr,KKdr,gca,VhalfCa,KCa,gkm,VhalfKm,KKm,VhalfNat,
                gkap,tauca,tstop,dur,amp,asyne,asyngmax,asynonset,asyntau,
               segkdr,segna3,segnap,seghd,segca,segim):
    
    soma.diam = diam
    soma.L = L # um
    soma.cm = cm*1.4884e-4/6.2832e-4
    soma.gl_hh = gl*1e-6
    soma.el_hh = el
    if w_Nat.value:
        soma.gbar_na3 = gnat*1e-3 
        soma.mhalf_na3 = VhalfNat
        soma.mk_na3 = KNat
        soma.segna3_na3 = segna3
    if w_Nap.value:
        soma.gbar_nap = gnap*1e-6
        soma.k_nap = KNap
        soma.vhalf_nap = VhalfNap
        soma.segnap_nap = segnap
    if w_H.value:
        soma.ghdbar_hd = gh*1e-6
        soma.kl_hd = KH
        soma.vhalfl_hd = VhalfH
        soma.seghd_hd = seghd
    if w_Kdr.value:
        soma.gbar_kdr = gkdr*1e-6
        soma.vhalfn_kdr = VhalfKdr
        soma.kn_kdr = KKdr
        soma.segkdr_kdr = segkdr
    if w_Cadyn.value:
        soma.gcabar_cadyn = gca*1e-6
        soma.vhalf_cadyn = VhalfCa
        soma.k_cadyn = KCa
        soma.segca_cadyn = segca
    #soma.gsAHPbar_sAHP = gsAHP*1e-6
    if w_Km.value:
        soma.gbar_im = gkm*1e-4
        soma.vhalf_im = VhalfKm
        soma.k_im = KKm
        soma.segim_im = segim
    if w_Kap.value:
        soma.gkabar_kap = gkap*1e-6
    if w_Capool.value:
        soma.taucas_capool = tauca
    
    if source == 1:
        Source = h.AlphaSynapse(soma(0.5))
        Source.e = asyne
        Source.gmax = asyngmax
        Source.onset = asynonset
        Source.tau = asyntau 
    if source == 0:
        Source = h.IClamp(soma(0.5))
        Source.delay = dur[0]
        Source.dur = dur[1]-dur[0]
        Source.amp = amp
    
    
    h.tstop = tstop
    h.v_init = el
    
    nc = h.NetCon(soma(0.5)._ref_v,None,sec=soma)
    nc.threshold = 0
    spvec = h.Vector()
    nc.record(spvec)
    passive = []
    h.run()
    
    if Source.amp < 0: 
        
        dt_V = v0_vec[int(dur[1])*40-10]-v0_vec[int(dur[0])*40-10]
        R_in = dt_V/Source.amp
        tau = soma.cm*R_in
        V_rest = v0_vec[int(tstop)*40-10]
        passive.append(R_in)
        passive.append(tau)
        passive.append(V_rest)
        print(dt_V)
    
  
      
    print('number of spikes:',len(spvec))
    print(passive,a.value,b.value,c.value)
    plt.figure(figsize=(12,8))
    plt.subplot(2,2,1)
    plt.plot(t_vec, v0_vec,'b')
    plt.xlim(0, tstop)
    plt.ylabel('mV')
    plt.subplot(2,2,2)
    if w_a.value:
        plt.plot(v0_vec,inf_nat,'k.',label = 'inf_nat')
    if w_b.value:
        plt.plot(v0_vec,inf_nap,'r.',label = 'inf_nap')
    if w_c.value:
        plt.plot(v0_vec,inf_h,'b.',label = 'inf_H')
    if w_d.value:
        plt.plot(v0_vec,inf_km,'y.',label = 'inf_km')
    if w_e.value:
        plt.plot(v0_vec,inf_kdr,'g.',label = 'inf_kdr')
    if w_f.value:
        plt.plot(v0_vec,inf_cadyn,'c.',label = 'inf_cadyn')
    plt.legend()
    plt.xlabel('voltage (mV)')
    plt.ylabel('Inf')
    plt.subplot(2,2,3)
    freq1 = []
    I1 = []
    if (Source.amp > 0): 
        for Source.amp in np.arange(amp,2.1,0.1):
            h.run()
            freq1.append(len(spvec)*1000/Source.dur)
            I1.append(Source.amp)
    plt.xlim(0,2)
    plt.ylim(0,200)
    plt.plot(I1,freq1,'yo',label = 'FIR')    
    for i in range(0, len(frequency)): 
        frequency[i] = float(frequency[i]) 
    for i in range(0, len(amplitude)): 
        amplitude[i] = float(amplitude[i])
    plt.plot(amplitude,frequency,'bo', label = 'Users FIR') 
    plt.legend()
    plt.subplot(2,2,4)   
    if w_a.value:
        plt.plot(v0_vec,tau_nat,'k.',label = 'Nat')
    if w_b.value:
        plt.plot(v0_vec,tau_nap,'r.',label = 'Nap')
    if w_c.value:
        plt.plot(v0_vec,tau_h,'b.',label = 'H')
    if w_d.value:
        plt.plot(v0_vec,tau_km,'y.',label = 'Km')
    if w_e.value:
        plt.plot(v0_vec,tau_kdr,'g.',label = 'Kdr')
    if w_f.value:
        plt.plot(v0_vec,tau_cadyn,'c.',label = 'Ca')
    plt.legend()
    plt.xlabel('voltage (mV)')
    plt.ylabel('Tau')
    plt.show()
    
clear_output()

```
